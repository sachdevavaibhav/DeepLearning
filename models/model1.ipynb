```python
from tensorflow import keras
from tensorflow.keras import layers
from data.load_data import load_data
from data.predict import extract_results
```

```python
(train_images, train_labels,test_images, test_labels) = load_data() #load the mnist dataset
```

```python
train_images.shape
```

```python
train_labels.shape
```

```python
model = keras.Sequential([
    layers.Dense(units=512, activation='relu'),
    layers.Dense(units=10, activation='softmax')
])
```

```python
model.compile(optimizer='rmsprop', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(train_images,train_labels, epochs=5, batch_size=128)
```

```python
test_images[0:12]
```

```python
predictions = model.predict(test_images)
predictions = extract_results(predictions)
predictions[:12]
```

```python
from data.predict import plot_results
plot_results(predictions, test_images, test_labels)
```
